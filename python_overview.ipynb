<br> Create a dictionary comprehension which iterates over the two given lists simultaniously and creats a dictionary where:
- keys are from list "a" and values are from list "b"<br>
- all values are squared <br>
- only even values of dictionary are kept <br>

The output will be this: {'b': 4, 'd': 16}

```python
#dictionaru comprehension
=["a","b","c","d","e"]
b=[1,2,3,4,5]

#a
d= {i:j for i,j in zip(a,b)}
print(d)
#B
s={key:value**2 for j in d for key,value in d.items()}
print(s)
#c
e={key:value for j in s for key,value in s.items() if value%2==0}
print(e)
```
